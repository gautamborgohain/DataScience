### Brief intro to data.table

Data.table is a package that extends the data frame functionalites and promises faster operations. The 'queryies' for data.table is like this : 

table_name[i,j,by]

where you place the subsetting rule at 'i'
 eg. table_name[1:5,]   -> selects 1 to 5th row
     or table_name[column_name %in% c("val1","val2")]   -> This selects the rows where 'column_name' contains values        Val1 and Val2 
     
     
At 'j' you do the calculation/manipulation 
 eg. if you want to select columns 'a' and 'b' as a data.table
     table_name[,.(a,b)] ->if.() is not used, then a vector will be returned
     
The main use of 'j' part is when it is used in conjuction with the' by' part

So say we want to calculate the the number of unique combinations of user_id and hour, we can use:
        sample_releasedata[,.N,by = list(user_id,hour)]
        again, if we want to rename the column to something like TOTAL_COUNT, we can use the .() to return the          data.table with the new column,  like :
        sample_releasdata [,.(TOTAL_COUNT = .N), by=  list(user_id,hour)]
        

Finally using all the parts of the data.table looks like this:

sample_releasedata[1:5,.(TOTAL_COUNT = .N),by = list(user_id,hour)]
This means first subset the sample_releasedata from 1 to 5, then group by user_id and hour and calculate the count of rows.
     

To have done the same operation using the aggregate function we would need to create an extra variable with all row values as 1 and we would aggregate on that row

```r
sample_release_copy = sample_release
sample_release_copy$x = 1
head(sample_release_copy)
houragg_copy = aggregate(sample_release_copy$x, by = list(sample_release_copy$user_id,sample_release_copy$hour), FUN = "sum")
head(houragg_copy)
nrow(houragg_copy)
nrow(houragg)
```

So now, lets do the cluster anlaysis part. We will start with the kmeans clustersing algorithm.

## K means Clustering

K means clustering is a sinple algorithm that works as follows:

1. Randomly select k number of cluster centroids
2. Calculate the distance between each data point and the centroid
3. Assign the data point to the centroid to which the distance is minimum
4. Recalculate the cluster center/centroid
5. Recalculate the distance between the data points and the new cluster centroids
6. Repeat till the cluster centroids do not change or till the number of iterations initially specified.


R provides a function to perform this procedure for us and we acn specify the parameters to tweak to our liking.


## Clustering on Hours

```r
hour_data = r_agg_hr[-1]
set.seed(1234)
fit = kmeans(hour_data,centers = 3,nstart = 15)
```

Now Lets visualize what the clusters we got

```r
library(fpc)
plotcluster(hour_data, fit$cluster, xlab = "Discriminant func 1", ylab = "Discriminant func 2")
```

We had just assumed 3 clusters would be good and asked R to give us data on 3 clusters, however, some people like to use the withinness score or the sum of squares with the groups to assess at what value of K is the clustering optimal, i.e the sum of squares minimal. The funtion below plots the 'scree plot' of the withiness score when the data is clustered in 2 to 7 clusters.

```r
wssplot = function(data, nc = 7, seed = 1234) {
  wss = (nrow(data) - 1) * sum(apply(data,2,var))
  for (i in 2:nc) {
    set.seed(seed)
    wss[i] = sum(kmeans(data, centers = i,nstart = 15)$withinss)
  }
  plot(1:nc, wss, type = "b", xlab = "Number of Clusters",ylab = "Within groups sum of squares")
}
```

```r
wssplot(hour_data,nc= 7)
```

This plot tells us that using 4 or 5 clusters maybe better than 3.

So lets form 5 clusters and visualize them

```r
set.seed(1234)
fit = kmeans(hour_data,centers = 5,nstart = 15)
plotcluster(hour_data, fit$cluster, xlab = "Discriminant func 1", ylab = "Discriminant func 2")
```

The set.seed function is to control the randomness. This allows for the results to be same


Also, we can do compare the clusters to a 2 component PCA plot

```r
clusplot(
    hour_data, fit$cluster, shade = T, lines = 0, color = T, lty = 4,main =
      'Principal Components plot showing K-means clusters'
  )
```

Now that we have got the clustering out of the way, Lets look at what these clusters actually represent. First, we would need map the clusters to the r_agg_hr data frame 

```r
r_agg_hr$cluster = fit$cluster
head(r_agg_hr)
```

Now that we have a data frame that has the information of what cluster each user_id belongs to ,we can look into which tweets are classified into which cluster. Inorder to do that, we need to merge the cluster data to the original data which has row wise information on tweets

```r
head(sample_releasedata)
```

```r
combineddf = merge(sample_releasedata,r_agg_hr, by = "user_id")
head(combineddf)
```

Now we have the data on each tweet and its cluster classification. We can get the count of these cluster associations to see based on the hour, what is the distribution or rather difference between the clusters.

To do that, we woud get the counts of each cluster in different data.tables like we did earlier and the combine them into one data.table.

```r
cluster1 = combineddf[cluster == 1,.(Cluster1 = .N), by= hour]
cluster2 = combineddf[cluster == 2,.(Cluster2 = .N), by= hour]
cluster3 = combineddf[cluster == 3,.(Cluster3 = .N), by= hour]
cluster4 = combineddf[cluster == 4,.(Cluster4 = .N), by= hour]
cluster5 = combineddf[cluster == 5,.(Cluster5 = .N), by= hour]

head(cluster1)

listdf = list(cluster1,cluster2,cluster3,cluster4,cluster5)
finaldf = Reduce(function(...) merge(..., all=T,by = 'hour'), listdf)
    
head(finaldf)
```

Now we are ready to plot the data by using ggplot2

```r
library(ggplot2)

finalmelt = melt(finaldf,id = 'hour')
names(finalmelt)[names(finalmelt)=="variable"] = "Clusters"
head(finalmelt)
```

ggplot2 requires the data to be in the long form, so we use reshape2 to melt the data frame

```r
  plt1 = ggplot(data=finalmelt,
                aes(x=hour, y=value, colour=Clusters)) +
    geom_line(size=2)+
    theme_bw()+
    xlab("Hours of a day")+
    ylab("Frequency of tweets")+
    ggtitle("Difference in clusters' time for tweeting - Hourly")+
    scale_x_continuous(breaks = seq(0,24,1))

plt1
```

This graph shows distribution of the clusters but it displays the count of tweets. A better way to understand the graph would be to calculate the percentages and see 'what percentage of the cluster1 tweets are at 12 o'clock' and so on

```r
finaldf[is.na(finaldf)] = 0
finaldf$Cluster.1 = finaldf$Cluster1/sum(finaldf$Cluster1) *100
finaldf$Cluster.2 = finaldf$Cluster2/sum(finaldf$Cluster2) *100
finaldf$Cluster.3 = finaldf$Cluster3/sum(finaldf$Cluster3) *100
finaldf$Cluster.4 = finaldf$Cluster4/sum(finaldf$Cluster4) *100
finaldf$Cluster.5 = finaldf$Cluster5/sum(finaldf$Cluster5) *100

finaldf[,c('Cluster1','Cluster2','Cluster3','Cluster4','Cluster5') := NULL][]
```

```r
finalmelt = melt(finaldf,id = 'hour')
names(finalmelt)[names(finalmelt)=="variable"] = "Clusters"

plt2 = ggplot(data=finalmelt,
                aes(x=hour, y=value, colour=Clusters)) +
    geom_line(size=2)+
    theme_bw()+
    xlab("Hours of a day")+
    ylab("Percentage of frequency of tweets %")+
    ggtitle("Difference in clusters' time for tweeting - Hourly - Percentage")+
    scale_x_continuous(breaks = seq(0,24,1))
plt2
```

So now it appears that twitter activity is high at night. Cluster.1's activity concentrates at around 12:00 pm. While Cluster.2 type users are pretty consistent in their use of twiiter and their usage is spread out. It can be seen however that the increase increases throughout the day Cluster.3 type users on the other hand do a bulk of their tweeting in the morning and then gradually slow down. CLuster.4 users conccentrate their tweeting later in the evening or night. Finally Cluster.5 users are quite similar to Cluster.2 however, Cluster.5 users start their day on Twitter earlier and do less tweeting at night compared to cluster.2.


Based on this analysis we can even classify the users in to different categories as such:

```r
names(finaldf)[names(finaldf)== "Cluster.1"]  = "Afternoon users"
names(finaldf)[names(finaldf)== "Cluster.2"]  = "Consistent - Late night"
names(finaldf)[names(finaldf)== "Cluster.3"]  = "Early morning heavy"
names(finaldf)[names(finaldf)== "Cluster.4"]  = "Evening users"
names(finaldf)[names(finaldf)== "Cluster.5"]  = "Consistent - Morning"

finalmelt = melt(finaldf,id = 'hour')
names(finalmelt)[names(finalmelt)=="variable"] = "Clusters"

plt2 = ggplot(data=finalmelt,
                aes(x=hour, y=value, colour=Clusters)) +
    geom_line(size=2)+
    theme_bw()+
    xlab("Hours of a day")+
    ylab("Percentage of frequency of tweets %")+
    ggtitle("Difference in clusters' time for tweeting - Hourly - Percentage")+
    scale_x_continuous(breaks = seq(0,24,1))
plt2
```

## Clustering on Week-days

Performing similar anlaysis on the weekdays:


```r
wday_data = r_agg_wday[-1]
wssplot(wday_data,nc= 7)
```

```r
set.seed(1234)
fit_wday = kmeans(wday_data,centers = 4,nstart = 15)
plotcluster(wday_data, fit_wday$cluster, xlab = "Discriminant func 1", ylab = "Discriminant func 2")
```

```r
r_agg_wday$cluster = fit_wday$cluster
combineddf_wday = merge(sample_releasedata,r_agg_wday, by = "user_id")

cluster1 = combineddf_wday[cluster == 1,.(Cluster1 = .N), by= wday]
cluster2 = combineddf_wday[cluster == 2,.(Cluster2 = .N), by= wday]
cluster3 = combineddf_wday[cluster == 3,.(Cluster3 = .N), by= wday]
cluster4 = combineddf_wday[cluster == 4,.(Cluster4 = .N), by= wday]

listdf_wday = list(cluster1,cluster2,cluster3,cluster4)
finaldf_wday = Reduce(function(...) merge(..., all=T,by = 'wday'), listdf_wday)

    
finaldf_wday[is.na(finaldf_wday)] = 0
finaldf_wday$Mid_Week_Weekend = finaldf_wday$Cluster1/sum(finaldf_wday$Cluster1) *100
finaldf_wday$Mid_week_heavy = finaldf_wday$Cluster2/sum(finaldf_wday$Cluster2) *100
finaldf_wday$All_week = finaldf_wday$Cluster3/sum(finaldf_wday$Cluster3) *100
finaldf_wday$Weekend_heavy= finaldf_wday$Cluster4/sum(finaldf_wday$Cluster4) *100


finaldf_wday[,c('Cluster1','Cluster2','Cluster3','Cluster4') := NULL]
    
finalmelt_wday = melt(finaldf_wday,id = 'wday')
names(finalmelt_wday)[names(finalmelt_wday)=="variable"] = "Clusters"

plt2_wday = ggplot(data=finalmelt_wday,
                aes(x=wday, y=value, colour=Clusters)) +
    geom_line(size=2)+
    theme_bw()+
    xlab("Days of a Week")+
    ylab("Percentage of frequency of tweets %")+
    ggtitle("Difference in clusters' time for tweeting - Weekly - Percentage")+
    scale_x_continuous(breaks = seq(0,6,1))
plt2_wday
```

Looking at the distribution of the clusters, we can notice the patterns of the clusters.
Cluster1 users are heavy on Sundays.
Cluster2 users are the heaviest users mid-week -Wednesday and Thursdays
Cluster3 users are on twitter the entire week!!
Cluster4 users are mid week and weekend users


## Cluster comparision


Now that we have two behaviours of twitter users, lets compare the two clusters by observing a contingency table

```r
tempdf = merge(r_agg_hr,r_agg_wday, by = 'user_id')
clusterdf = data.frame(ClusterByWeek = tempdf$cluster.y,ClusterByHours = tempdf$cluster.x)

clusterdf$ClusterByHours = as.numeric(clusterdf$ClusterByHours)
clusterdf$ClusterByWeek = as.numeric(clusterdf$ClusterByWeek)

clusterdf$ClusterByHours[clusterdf$ClusterByHours == 1] <- "Afternoon users"
clusterdf$ClusterByHours[clusterdf$ClusterByHours == 2] <- "Consistent - Late night"
clusterdf$ClusterByHours[clusterdf$ClusterByHours == 3] <- "Early morning heavy"
clusterdf$ClusterByHours[clusterdf$ClusterByHours == 4] <- "Evening users"
clusterdf$ClusterByHours[clusterdf$ClusterByHours == 5] <- "Consistent - Morning"

clusterdf$ClusterByWeek[clusterdf$ClusterByWeek == 1] <- "Weekend_heavy "
clusterdf$ClusterByWeek[clusterdf$ClusterByWeek == 2] <- "Mid_week_heavy"
clusterdf$ClusterByWeek[clusterdf$ClusterByWeek == 3] <- "All_week"
clusterdf$ClusterByWeek[clusterdf$ClusterByWeek == 4] <- "Mid_Week_Weekend"

library(gridExtra)
grid.arrange(plt2,plt2_wday)

table(clusterdf)
```

Finally, by observing the contingency table, we can see that the All_week users are more likely to be using Twitter consistently all day with more activity in the morning. Where as Mid_week_heavy  users are also likely all day users with more focus at night.
Lastly weekend heavy users tend to use Twitter in the afternoon and late night.
